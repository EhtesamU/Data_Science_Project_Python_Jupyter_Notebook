# Data cleaning
This cleaning file is created to join and clean data for happiness 2015-2017 and 2018-2019

```python
import pandas as pd
import clean
```

## 2015 - 2017

```python
#load 2015 data
load_happy15 = pd.read_csv('../data/raw-data/Happiness-Score/2015.csv')
happy15 = load_happy15.drop(['Happiness Rank','Standard Error'],axis=1)

# rename col
happy15.rename(columns={'Happiness Score':'happiness_score','Economy (GDP per Capita)':'GDP_per_cap','Health (Life Expectancy)':'life_expect','Trust (Government Corruption)':'gov_trust','Dystopia Residual':'dystopia_residual'},inplace=True)

# create continent col
clean.create_continent(happy15) 

# adding year col
happy15['Year'] = "2015"

happy15.head()
```

```python
#load 2016 data
load_happy16 = pd.read_csv('../data/raw-data/Happiness-Score/2016.csv')
happy16 = load_happy16.drop(['Happiness Rank','Lower Confidence Interval', 'Upper Confidence Interval'],axis=1)

# rename col
happy16.rename(columns={'Happiness Score':'happiness_score','Economy (GDP per Capita)':'GDP_per_cap','Health (Life Expectancy)':'life_expect','Trust (Government Corruption)':'gov_trust','Dystopia Residual':'dystopia_residual'},inplace=True)

# create continent col
clean.create_continent(happy16)

# adding year
happy16['Year'] = "2016"

happy16.head()
```

```python
# load 2017 data
load_happy17 = pd.read_csv('../data/raw-data/Happiness-Score/2017.csv')
happy17 = load_happy17.drop(['Happiness.Rank','Whisker.high', 'Whisker.low'],axis=1)

# rename col
happy17.rename(columns={'Happiness.Score':'happiness_score','Economy..GDP.per.Capita.':'GDP_per_cap','Health..Life.Expectancy.':'life_expect','Trust..Government.Corruption.':'gov_trust','Dystopia.Residual':'dystopia_residual'},inplace=True)

happy17.head()
```

```python
# clean hongkong and taiwan cell
clean.hongkong(happy17,False)
clean.taiwan(happy17,False)

# adding region and continent for year 2017
clean.create_region(happy16,happy17)
clean.create_continent(happy17)

# adding year for 2017
happy17['Year'] = "2017"

happy17.head()
```

```python
# re arrange all col for appending
clean.arrange_col_15_17(happy15)
clean.arrange_col_15_17(happy16)
clean.arrange_col_15_17(happy17)

# appending 2015, 2016, 2017
happy1517 = happy15.append(happy16,ignore_index=True).append(happy17,ignore_index=True)
```

```python
clean.delete_less_than_record(happy1517,3)
```

```python
happy1517.to_csv('../data/clean-data/clean-happiness-15-17_using python.csv', index=False)  
```

## 2018 - 2019

```python
#load 2018 data
load_happy18 = pd.read_csv('../data/raw-data/Happiness-Score/2018.csv')
load_happy18.head()
happy18 = load_happy18.drop(['Overall rank'],axis=1)

# rename col
happy18.rename(columns={'Country or region':'Country','Score': 'happiness_score','GDP per capita':'GDP_per_cap','Social support':'social_support','Healthy life expectancy':'life_expect','Freedom to make life choices': 'Freedom','Perceptions of corruption':'corruption_perceptions'},inplace=True)

# adding year col
happy18['Year'] = "2018"

# clean hongkong and taiwan cell
clean.hongkong(happy18,False)
clean.taiwan(happy18,False)

happy18.head()  
```

```python
#load 2019 data
load_happy19 = pd.read_csv('../data/raw-data/Happiness-Score/2019.csv')
load_happy19.head()
happy19 = load_happy19.drop(['Overall rank'],axis=1)

# rename col
happy19.rename(columns={'Country or region':'Country','Score': 'happiness_score','GDP per capita':'GDP_per_cap','Social support':'social_support','Healthy life expectancy':'life_expect','Freedom to make life choices': 'Freedom','Perceptions of corruption':'corruption_perceptions'},inplace=True)

# adding year col
happy19['Year'] = "2019"


# clean hongkong and taiwan cell
clean.hongkong(happy19,False)
clean.taiwan(happy19,False)

happy19.head()
```

```python
# arrange col for appending
clean.arrange_col_18_19(happy18)
clean.arrange_col_18_19(happy19)

happy1819 = happy18.append(happy19,ignore_index=True)
```

```python
# removing countries with missing record
clean.delete_less_than_record(happy1819,2)
```

```python
happy1819.to_csv('../data/clean-data/clean-happiness-18-19_using python.csv', index=False)  
```
